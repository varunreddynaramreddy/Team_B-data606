#### Outlier capping logic
- Normal variables: mu, 3*sigma rule
- skewed distribution: Inter-Quartile Range (IQR) proximity rule
- Constant variables: no outlier treatment

```python
### Outlier treatment based on graphical observation of distribution and outliers

outlierDict = {'age': 'Normal', 'balance': 'skewed', 'day': 'constant', 'duration': 'skewed', 'campaign': 'skewed'}
```

```python
outlierDict
```

```python
def outlierTreatment(var, dist):
    '''
    Methods take pandas series and distribution of series as input and return capped series based on distribution provided
    
    '''
    if dist == 'skewed':
        percentile25 = var.quantile(0.25)
        percentile75 = var.quantile(0.75)
        upper_limit = percentile75 + 3 * (percentile75-percentile25)
        lower_limit = percentile25 - 3 * (percentile75-percentile25)
        
        return np.where(var > upper_limit, upper_limit, np.where(var < lower_limit,lower_limit,var))
    elif dist == 'Normal':
        upper_limit = np.mean(var) + 3 * np.std(var)
        lower_limit = np.mean(var) - 3 * np.std(var)
        return np.where(var > upper_limit, upper_limit, np.where(var < lower_limit,lower_limit,var))           
    else:
        return var
        
```

```python
for i,j in zip(outlierDict.keys(), outlierDict.values()):
    rawData[i+'_OT'] = outlierTreatment(rawData[i], j)
```
